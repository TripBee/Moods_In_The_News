```python
## Ok! so our goal today is to create a script that vizualizes the recent Mood in Twitter News
## The first step is to grab some tweets from Twitter's API
## Today we are the using tweepy library.
import tweepy

## In order to request information from twitter we need to authenticate ourselves as a legitimate person.
## We load our ID codes...
from config import (consumer_key, 
                    consumer_secret, 
                    access_token, 
                    access_token_secret)

## ... and then authenticate ourselves.
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())
## We can request information from twitter now!

## Then we analyze the tweets to see how positive and negetive they are.
## The VADER Sentiment Intensity Analyzer can do this.
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

## The next step is to format the data into tables that can be read by graphing software.
## These libraries should get the job done!
import numpy as np
import pandas as pd

## Now we need graphing software that visualizes the results.
## These should work.
from datetime import datetime
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')

## And one more library for formatting the output,
import os

## Here We Go!
```

```python
# What accounts are we looking at today?
accounts = ("@BBC", "@CBS", "@CNN", "@FoxNews", "@nytimes")
```

```python
# create an array to hold the results of our querry 
sentiments= []

# loop through each of the accounts in the accounts list
for account in accounts:
    # set a counter for how many tweets we have looked at in each account
    counter = 1
    # Variable for max_id (this will tell the API how deep into the timeline to go when pulling each of the pages of tweets)
    oldest_tweet = None
    # Loop through 5 pages of tweets (total 100 tweets)
    for x in range(5):
        # Get all tweets from home feed
        public_tweets = api.user_timeline(account, max_id = oldest_tweet)
        # Loop through all tweets 
        for tweet in public_tweets:   
            # Run Vader Analysis on each tweet
            results = analyzer.polarity_scores(tweet["text"])
            # name the results we care about  
            compound = results["compound"]
            pos = results["pos"]
            neu = results["neu"]
            neg = results["neg"]
            # Add sentiments for each tweet into a list
            sentiments.append({
                "User": account,
                "Date": tweet["created_at"],
                "Compound": compound,
                "Positive": pos,
                "Negative": neu,
                "Neutral": neg,
                "Tweets Ago": counter,
                "Message": tweet["text"]
                                })
            # save the ID of the last tweet so the next page will pick up where we left off           
            oldest_tweet = tweet["id"] - 1
        # Add to counter 
            counter += 1  
```

```python
# Convert sentiments to DataFrame
df = pd.DataFrame.from_dict(sentiments)
df.count()
```

```python
# Create plot
x_vals = df["Tweets Ago"]
y_vals = df["Compound"]
plt.scatter(x_vals, y_vals, marker="o", c=["g","r","b","y","w"], alpha=0.7)

# Incorporate the other graph properties
now = datetime.now()
now = now.strftime("%Y-%m-%d  at  %H:%M")
plt.title(f"Sentiment Analysis of  \
{accounts[0]}, {accounts[1]}, {accounts[2]}, {accounts[3]}, and {accounts[4]} Tweets \
    on  {now}")
plt.ylabel("NEG<---Tweet Polarity--->POS")
plt.xlabel("Newer<------Tweets Ago----->Older")
if not os.path.exists("Output"):
    os.makedirs("Output") 
if not os.path.exists("Output/twitter_sentiment_scatter.png"):
    plt.savefig("Output/twitter_sentiment_scatter.png")
plt.show()
```
